```python
from scipy import optimize
import matplotlib.pyplot as plt
import numpy as np
import cv2 as cv
import random
import glob
```

```python
img = plt.imread(r'C:\Users\admin\Desktop\projects\1\pngs\pp.png')
```

```python
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
plt.imshow(gray)
```

```python
gray = (gray*255).astype(np.uint8)
edges = cv.Canny(gray,100,200)
plt.imshow(edges)
```

```python
plist = np.argwhere(edges == 255)
```

```python
path = r'C:\Users\admin\Desktop\projects\1\pngs\psyvid\\'
frame = 1
for ite in range(50):
    px1 = []
    py1 = []
    ran = random.uniform(0, 1)
    for arr in plist:
        if random.uniform(0, 1)>ran:
            px1.append(arr[0])
            py1.append(arr[1])
            ran = random.uniform(0, 1)
    plt.plot(px1,py1)
    plt.savefig(path+'i'+str(frame)+'.jpg')
    frame = frame+1
```
