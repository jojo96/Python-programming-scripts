
**HADAMARD Matrix**


Python **recursive** function
---
Write a program which takes an input N and outputs an NxN Hadamard pattern where N=2^n where n is some integer. The Hadamard function should be defined recursively. Assume N<=256. 
---
A Hadamard pattern is a black and white pattern defined recursively: A 1x1 Hadamard pattern is a single black square. 2Nx2N Hadamard pattern is obtained by aligning 4 copies of the NxN pattern in the form of a 2x2 grid and then inverting the colors of all the squares in the lower right NxN copy.

---
Input Format: program should take one integer N. assume N is a power of 2. 

---



Output Format: the NxN pattern should be printed in N rows of N columns, with columns separated by space and rows separated by a newline. Output 1 for black square and 0 for white square

```
import numpy as np


def hadamard(n):
  mat=[]
  if n==2:
    mat=[[0,0],[0,1]]
    return np.array(mat)
  m1=hadamard(n/2)
  m2=hadamard(n/2)
  m3=hadamard(n/2)
  m4=1-hadamard(n/2)
  return np.vstack((np.hstack((m1,m2)),np.hstack((m3,m4))))
```
